## Step 1: Load the Dataset

```python
import pandas as pd

df = pd.read_csv('Spam_dataset/spam.csv', encoding='latin-1')

df.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)

df = df[['label', 'text']]

df
```

## Step 2: Apply TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data and creates the raw feature matrix
X = tfidf_vectorizer.fit_transform(df['text'])
y = df['label']

print(f"TF-IDF Matrix Shape (Original): {X.shape}")
```

## Step 3 & 4: Feature Selection & Reporting Removed Features

```python
from sklearn.feature_selection import VarianceThreshold

# 1. Setup the selector with the requested threshold
selector = VarianceThreshold(threshold=0.1)

try:
    # 2. Try to fit and transform
    X_selected = selector.fit_transform(X)
    
    # If successful, calculate how many were removed
    removed_features = X.shape[1] - X_selected.shape[1]
    print(f"Threshold 0.1 applied successfully.")

except ValueError:
    # 3. If ValueError occurs (all features removed), catch it
    print("Threshold 0.1 was too high (all features would be removed).")
    print("Falling back to keeping all features.\n")
    
    X_selected = X
    removed_features = 0

# 4. Report results
print(f"Original Features: {X.shape[1]}")
print(f"Selected Features: {X_selected.shape[1]}")
print(f"Features Removed:  {removed_features}")
```

## Step 5: Stratified Hold-out

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, 
    y, 
    test_size=0.3, # 70% train, 30% test
    stratify=y, # matches the class proportions (no shuffle)
    random_state=1234, # guarantees the same split every time
    shuffle=True
)

print("Stratified hold-out split completed.")
```

## Step 6: Report Matrix Shapes

```python
print(f"Training Matrix Shape: {X_train.shape}")
print(f"Testing Matrix Shape:  {X_test.shape}")
```

```python
print("--- Training Set Class Distribution ---")
print(y_train.value_counts())
print("\nRatio:")
print(y_train.value_counts(normalize=True))
```

```python
print("\n--- Testing Set Class Distribution ---")
print(y_test.value_counts())
print("\nRatio:")
print(y_test.value_counts(normalize=True))
```

## Step 7: Top 10 and Bottom 10 Rows

```python
X_train_df = pd.DataFrame(X_train.toarray())

print("\n===== REPORT (7) Top 10 Rows (X_train) =====")
display(X_train_df.head(10))
```

```python
print("\n===== REPORT (7) Bottom 10 Rows (X_train) =====")
display(X_train_df.tail(10))
```

## Step 8: Classification Matrix

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize and train Naive Bayes
clf = MultinomialNB()
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluation
acc = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {acc:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
